```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
Data_path = "Resources/Crash_Data.csv"

Accident_data = pd.read_csv(Data_path,low_memory=False)

Accident_data.head(10)
```

```python
Accident = Accident_data.loc[(Accident_data["Year"] >= 2018) & (Accident_data["Year"] <= 2021)] 

Accident.head()
```

```python
import seaborn as sns
#Cleaning data, check how many blank data in the set
fig, ax = plt.subplots(figsize=(10,5))     
cols = Accident.columns 
colours = ['#FFB2A2', '#00008f'] 
sns.heatmap(Accident_data[cols].isnull(), cmap=sns.color_palette(colours))
ax.set_xticklabels(ax.get_xticklabels(),rotation =90)
plt.title("Visualizing the Nothing in Dataset of Australian Fatal Road Accident 2018-2021", bbox={'facecolor':'0.8', 'pad':5})

# Save graph and show the grap
plt.savefig("Images/Visualizing the Nothing.png")
plt.show()
```

```python
#Does the age of the driver affect the accident?
Age_DATAset=Accident.groupby(['Age'])["Crash ID"].count().to_frame()
#Age_DATAset=Age_DATAset.reset_index()

#pd.set_option('display.max_rows',None)
Age_DATAset
```

```python
#set the figure title
plt.title("Number of accidents occurring in each age range in Australian Fatal Road Accident 2018-2021", bbox={'facecolor':'0.8', 'pad':5})

# Plot as a line chart
Age, = plt.plot(Age_DATAset, color="Red", label="Number of Accidents" )

# Create a legend for our chart
plt.legend(handles=[Age], loc="best")
plt.xlabel('Age')
plt.ylabel('Number of car accidents')


# Save graph and show the grap
plt.savefig("Images/AgeDriverinCarCrash.png")
plt.show()
```

```python
#Hypothesis testing

#set the figure title
plt.title("Number of accidents occurring in each age range in Australian Fatal Road Accident 2018-2021", bbox={'facecolor':'0.8', 'pad':5})

Age=Age_DATAset.reset_index()
plt.scatter(Age['Age'],Age['Crash ID'])
clb = plt.colorbar()
plt.xlabel("Age")
plt.ylabel("Crash number")

# Save graph and show the grap
plt.savefig("Images/AgeScatterAccidents.png")
plt.show()
```

```python

quartiles = Age['Crash ID'].quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of accidents is: {lowerq}")
print(f"The upper quartile of accidents is: {upperq}")
print(f"The interquartile range of accidents is: {iqr}")
print(f"The the median of accidents is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")

outlier_occupancy = Age.loc[(Age['Crash ID'] < lower_bound) | (Age['Crash ID'] > upper_bound)]
outlier_occupancy
```

```python
# Add the linear regression equation and line to plot
from scipy.stats import linregress

plt.title("Number of accidents occurring in each age range in Australian Fatal Road Accident 2018-2021", bbox={'facecolor':'0.8', 'pad':5})

x_values = Age['Age']
y_values = Age['Crash ID']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Age')
plt.ylabel('Number of car accident')

# Save graph and show the grap
plt.savefig("Images/AgelinearAccidents.png")
plt.show()
```
